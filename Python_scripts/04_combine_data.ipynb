```python
# load in all three data sets
transaction_data = pd.read_csv('transaction_data_cleaned.csv')
customer_demographic = pd.read_csv('customer_demographic_cleaned_data.csv')
customer_address = pd.read_csv('customer_address_cleaned_data.csv')
```

```python
# merge all three data together
transaction_demographic = pd.merge(transaction_data,customer_demographic , on = 'customer_id', how = 'inner')
all_three_data = pd.merge(transaction_demographic, customer_address, on = 'customer_id', how = 'inner')

# print few lines
all_three_data.head()
```
